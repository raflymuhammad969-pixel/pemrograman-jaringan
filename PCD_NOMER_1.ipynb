```python
# ============================
# Soal No.1 - Transformasi Citra Grayscale
# CPMK 1 dan CPMK 2
# ============================

import cv2
import numpy as np
import matplotlib.pyplot as plt

# --- 1. Baca citra grayscale ---
# Ganti 'gambar.jpg' dengan nama file kamu (bisa .jpg atau .png)
# Contoh: img = cv2.imread('pemandangan.png', cv2.IMREAD_GRAYSCALE)
img = cv2.imread('sate ayam.jpg', cv2.IMREAD_GRAYSCALE)

if img is None:
    raise FileNotFoundError("Gambar tidak ditemukan! Pastikan file ada di folder yang sama.")

# --- 2. Transformasi Negatif ---
negatif = 255 - img

# --- 3. Transformasi Logaritmik ---
# Rumus: G = c * log(1 + F)
# c = 255 / log(1 + max(F))
c = 255 / np.log(1 + np.max(img))
log_transform = c * np.log(1 + img)
log_transform = np.array(log_transform, dtype=np.uint8)

# --- 4. Tampilkan hasil ---
plt.figure(figsize=(12, 8))

plt.subplot(3, 2, 1)
plt.imshow(img, cmap='gray')
plt.title("Citra Asli")
plt.axis('off')

plt.subplot(3, 2, 2)
plt.hist(img.ravel(), bins=256)
plt.title("Histogram Citra Asli")

plt.subplot(3, 2, 3)
plt.imshow(negatif, cmap='gray')
plt.title("Citra Negatif")
plt.axis('off')

plt.subplot(3, 2, 4)
plt.hist(negatif.ravel(), bins=256)
plt.title("Histogram Citra Negatif")

plt.subplot(3, 2, 5)
plt.imshow(log_transform, cmap='gray')
plt.title("Citra Logaritmik")
plt.axis('off')

plt.subplot(3, 2, 6)
plt.hist(log_transform.ravel(), bins=256)
plt.title("Histogram Citra Logaritmik")

plt.tight_layout()
plt.show()

# --- 5. Analisis Statistik Sederhana ---
print("=== Statistik Citra ===")
print(f"Original -> min: {img.min()}, max: {img.max()}, mean: {img.mean():.2f}, std: {img.std():.2f}")
print(f"Negatif  -> min: {negatif.min()}, max: {negatif.max()}, mean: {negatif.mean():.2f}, std: {negatif.std():.2f}")
print(f"Log      -> min: {log_transform.min()}, max: {log_transform.max()}, mean: {log_transform.mean():.2f}, std: {log_transform.std():.2f}")
print(f"Nilai konstanta c = {c:.3f}")
```
